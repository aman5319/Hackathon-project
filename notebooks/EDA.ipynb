# Date Features

```python
def week_of_month(tgtdate):
    import datetime
    import calendar
    days_this_month = calendar.mdays[tgtdate.month]
    for i in range(1, days_this_month):
        d = pd.datetime(tgtdate.year, tgtdate.month, i)
        #Change to isoweek format which assigns 1 to Monday
        # d.isoweekday() is same as (d.weekday()+1)
        if d.day - (d.weekday()+1) >= 0:
            startdate = d
            #print("s",startdate)
            break
    return (tgtdate - startdate).days //7 + 1
```

```python
def add_datepart(df, fldname, drop=False, time=False):
    
    import re
    fld = df[fldname]
    fld_dtype = fld.dtype
    if isinstance(fld_dtype, pd.core.dtypes.dtypes.DatetimeTZDtype):
        fld_dtype = np.datetime64

    if not np.issubdtype(fld_dtype, np.datetime64):
        df[fldname] = fld = pd.to_datetime(fld, infer_datetime_format=True)
    targ_pre = re.sub('[Dd]ate$', '', fldname)
    attr = ['Year', 'Month', 'Week', 'Day', 'Dayofweek', 'Dayofyear',
            'Is_month_end', 'Is_month_start', 'Is_quarter_end', 'Is_quarter_start', 'Is_year_end', 'Is_year_start']
    if time: attr = attr + ['Hour', 'Minute', 'Second']
    for n in attr: df[targ_pre + n] = getattr(fld.dt, n.lower())
    #Pandas uses Monday-0 ,Sunday-6.To make in accordance to isoweek (Monday-1 ,Sunday-7)
    df[targ_pre + 'Dayofweek'] = df[targ_pre + 'Dayofweek']+1 
    df[targ_pre + 'Week_Num'] = [week_of_month(row[1]) for row in pd.DataFrame(fld).itertuples()]
    df[targ_pre + 'Elapsed'] = fld.astype(np.int64) // 10 ** 9
    attr_dayflag=["Is_Monday","Is_Tuesday","Is_Wednesday","Is_Thursday","Is_Friday"]
    for i, n in enumerate(attr_dayflag,1):df[targ_pre+n]=df[targ_pre + 'Dayofweek']==i
    attr_weekflag=["Is_Week0","Is_Week1","Is_Week2","Is_Week3","Is_Week4","Is_Week5"]
    for i, n in enumerate(attr_weekflag,0):df[targ_pre+n]=df[targ_pre + 'Week_Num']==i
    #for n in attr_flag:df.drop([targ_pre+n],axis=1,inplace=True)
    if drop: df.drop(fldname, axis=1, inplace=True)

def is_date(x): return np.issubdtype(x.dtype, np.datetime64)
```

```python
def create_cols(df,colnames):
    for i in colnames:
        df[i] = 0
    return df
```

```python
def extract_values(rows,total_cols):
    categories = rows["cat"]
    subcategories = rows['sub_cat']
    cat_subcat = categories + subcategories
    for i in cat_subcat:
        if i in total_cols:
            rows[i] = rows [i] +1
    return rows
```
